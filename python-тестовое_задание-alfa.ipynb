```python
import numpy as np
import pandas as pd
import random
```

```python
# Необходимо случайно сгенерировать список целочисленных значений из диапазона от 1 до 100 включительно. 
# Длина списка 36 символов.

l = np.random.randint(1, 101, size=36)
l
```

```python
# Преобразовать список в квадратную матрицу. Заполнять матрицу элементами списка построчно.

m = np.reshape(l, (6,6))
m
```

```python
# Возвести в квадрат значения элементов, расположенных выше главной диагонали матрицы.

n = np.square(m.copy(), out=m.copy(), where=np.triu(np.ones((6, 6), dtype=bool), k=1))
n
```

```python
# Для элементов, расположенных ниже побочной диагонали, увеличить на 2 значения, если i+j кратно трем.

i_plus_j = np.add(*np.ogrid[:6, :6])

k = np.where((i_plus_j > 5) & (i_plus_j % 3 == 0), n+2, n)
k
```

```python
# Преобразовать матрицу в датафрейм, название столбцов датафрейма должно быть в формате «Стобец № j».

df1 = pd.DataFrame(k, columns=[f"Стобец №{j}" for j in range(0, np.shape(k)[1])])
df1
```

```python
# Создать ещё один датафрейм с размерностью и названием полей, аналогичным датафрейму из пункта 5. 
# Заполнить датафрейм рандомными числами от 100 до 200.

h = np.random.randint(100, 201, size=(6, 6))

df2 = pd.DataFrame(h, columns=[f"Стобец №{j}" for j in range(0, np.shape(h)[1])])
df2
```

```python
# Объединить датафреймы из пункта 5 и 6 вертикально.

df = df1.append(df2, ignore_index=True)
df
```

```python
# В объединенном датафрейме создать столбец «Категория суммы», которое принимает значения 
# «Сумма от 0-100», «Сумма 100-1000», «Сумма 1000-10000», «Сумма 10000+» на основании построчной суммы элементов.

def categorize(x):
    if x < 100:
        return "Сумма 0-99"
    elif x < 1000:
        return "Сумма 100-999"
    elif x < 10_000:
        return "Сумма 1000-9999"
    else:
        return "Сумма 10000+"


df["Сумма"] = df.sum(axis=1)
df["Категория суммы"] = df.Сумма.apply(categorize)
df
```

```python
# Вывести на экран 25, 50, 75 персентиль по категориям суммы (используя группировку).

df[["Сумма", "Категория суммы"]].groupby("Категория суммы").quantile((.25, .5, .75))
```
